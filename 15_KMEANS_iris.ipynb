```python
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt0327

iris0327 = load_iris()
print(iris0327.keys())
print(iris0327.data.shape)
print(iris0327.feature_names)
print(iris0327.target_names)
print("此鳶尾花資料集的表格敘述:", iris0327.DESCR)
print("鳶尾花資料集之第一個資料data內容:",iris0327.data[0])
print("鳶尾花資料集之第一個資料的預測目標target:", iris0327.target[0])
```

```python
model0327 = KMeans(n_clusters=3, random_state=1)
model0327.fit(iris0327.data)
print(model0327.labels_)
print(model0327.cluster_centers_)
```

```python
# 利用原本資料集之key來繪製散布圖
scatter = plt0327.scatter(iris0327.data[:,2], iris0327.data[:,3], c=iris0327.target)
plt0327.xlabel('Petal_length')
plt0327.ylabel('Petal_width')
plt0327.legend( handles= scatter.legend_elements()[0], labels= iris0327.target_names.tolist())
plt0327.show()
```

```python
# 利用KMeans分類(群)後的標記來繪製散布圖scatter = plt0327.scatter(iris0327.data[:,2], iris0327.data[:,3], c=model0327.labels_)
plt0327.xlabel('Petal_length')
plt0327.ylabel('Petal_width')
```

```python
# 利用特別的標記(星號)標示各類(群)的Center
plt0327.scatter(model0327.cluster_centers_[:,2], model0327.cluster_centers_[:,3] ,marker='*', c='blue' , s=150)
plt0327.show()
```
