```python
import numpy as npy
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential
from tensorflow.keras.optimizers import SGD
x = npy.array([[0.,0.],[0.,1.],[1.,0.],[1.,1.]])
y = npy.array([0.,1.,1.,0.])
model0601 = Sequential() #使用Sequential來串接順序模型model0601的後續輸入等同於Sequential(Dense(........))
model0601.add(Dense(units=4, input_shape=(2,), kernel_initializer='normal', activation='relu'))
model0601.add(Dense(units=1, activation='sigmoid'))
model0601.compile(loss='binary_crossentropy', optimizer=SGD(learning_rate=0.01), metrics=['accuracy'])
print(model0601.summary())
history0601 = model0601.fit(x, y, epochs=10000, batch_size=4, verbose=0)
print(model0601.predict(x))
```

```python
from tensorflow.keras.models import load_model
import numpy as npy
model0601.save('model_XOR0601.keras')
model0601_NEW = load_model('model_XOR0601.keras')
x = npy.array([[0.,1.], [0.,0.], [1.,0.], [1.,1.]])
y_new = model0601_NEW.predict(x)
print(y_new)
```

```python
import matplotlib.pyplot as plt0601
plt0601.plot(history0601.history['accuracy'])
plt0601.title('Model0601 Correct rate')
plt0601.ylabel('Correct rate')
plt0601.xlabel('Epoch')
plt0601.show()
plt0601.plot(history0601.history['loss'])
plt0601.title('Model0601 loss')
plt0601.ylabel('loss')
plt0601.xlabel('Epoch')
plt0601.show()
```
